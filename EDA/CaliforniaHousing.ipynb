### Create a Test Set through stratified random sampling on the income variable

**Why a test set ?** <br>
**Why stratify the data on the income variable before generating a test set ?** <br>

### *``Your answers here``*

<span style='color:red'> Why a test set?
<span style='color:red'>Creating a test set is a fundamental step in any machine learning project. It serves multiple purposes:

<span style='color:red'>- Model Evaluation: A test set allows you to assess how well your chosen model will perform on unseen data.
<span style='color:red'>- Avoid Overfitting: The test set helps in detecting overfitting, a modeling error that occurs when a machine learning model performs well on the training data but poorly on new, unseen data.
<span style='color:red'>- Reproducibility: Using a fixed test set (often with a set random seed, like 42 in this case) helps in comparing different models under the same conditions, making the comparisons fair and the results reproducible.




<span style='color:red'>Why stratify the data on the income variable before generating a test set?
<span style='color:red'>Stratifying the data on a particular variable, such as income in this case, ensures that the test set is representative of the various categories of the income in the whole dataset. This is crucial for several reasons:

<span style='color:red'>- Representative Sampling: Stratification ensures that each stratum (or category) of the income variable is adequately represented in both the training and the test set.
<span style='color:red'>Reduced Bias: Stratification can reduce sampling bias, which may occur if the test set is not representative of the overall population.
<span style='color:red'>- Improved Accuracy: If income is an important variable for predicting the target variable, stratification ensures that the test set includes all categories of income, potentially leading to a more accurate evaluation of the model.
<span style='color:red'>- Reproducibility: Using a random seed (like 42) along with stratification ensures that the sampling is reproducible, thereby making it easier to validate the results or compare with other models.</span>

### 2.1.2 - Case where you have missing values in several variables

Let us now build a dataset with multiple missing values : <br>
Start with a function generating missing values in a chosen column of a dataframe

```python
def col_miss(df, col, max_miss):
    """
    df : a pandas dataframe
    col : the name of the variable column
    max_miss : the maximum number of missing values
    returns a data frame with a random number of missing values on col
    """

    df_copy = df.copy()

    num_miss = random.randint(0, max_miss)

    miss = np.random.choice(df.index, num_miss, replace=False)

    df_copy.loc[miss, col] = None

    return df_copy[col]
```

In theory, each of these has its advantages and disadvantages.</br>
For SimpleImputer.</br>
1. computational efficiency: SimpleImputer is very fast and efficient, especially for large data sets.

4. complexity and interpretability: although KNNImputer may be more accurate, it is not as interpretable as SimpleImputer.</br>

### 2.2.2 - Building a Pipeline

Let us start with a custom transformer to be used to add attributes

```python
housing_df.values
```

```python
from sklearn.base import BaseEstimator, TransformerMixin

# column index
rooms_ix, bedrooms_ix, population_ix, households_ix = 3, 4, 5, 6


class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, add_bedrooms_per_room=True):  # no *args or **kargs
        self.add_bedrooms_per_room = add_bedrooms_per_room

    def fit(self, X, y=None):
        return self  # nothing else to do

    def transform(self, X):
        rooms_per_household = X[:, rooms_ix] / X[:, households_ix]
        population_per_household = X[:, population_ix] / X[:, households_ix]
        if self.add_bedrooms_per_room:
            bedrooms_per_room = X[:, bedrooms_ix] / X[:, rooms_ix]
            return np.c_[
                X, rooms_per_household, population_per_household, bedrooms_per_room
            ]
        else:
            return np.c_[X, rooms_per_household, population_per_household]
```

```python
attr_adder = CombinedAttributesAdder(add_bedrooms_per_room=False)
houses_plus = attr_adder.transform(housing_df.values)
```

## 3.2 - Estimating the models with *cross validation*

**Explain what the following function accomplishes**

```python
def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())
```

### *``This display_scores function accepts as input an array of values scores, typically the results of a model performance assessment metric (such as MSE, MAE, or accuracy) over multiple cross-validation runs. The function performs the following three main tasks:``*

1. Print the score itself
2. Compute the mean value and print it
3. Compute the standard deviation and print it

#### Linear Regression

```python
lr = LinearRegression()
```

```python
from sklearn.model_selection import cross_val_score

lr_scores = cross_val_score(
    lr, X, y, scoring="neg_mean_squared_error", cv=10
)  #

# Since the scores are negative, we make them positive and then take square root to get RMSE
lr_rmse = np.sqrt(-lr_scores)

display_scores(lr_rmse)
```

## 3.3 - Tuning the model with Grid Search and Randomized Search

#### Example : Random Forest

```python
from sklearn.model_selection import GridSearchCV
```

```python
rf_grid = {"n_estimators": [30, 60, 100], "max_features": [8, 10, 15]}

# Initialize the Random Forset Model
rf = RandomForestRegressor(random_state=42)

# Initialize the grid reserch
grid_search = GridSearchCV(
    estimator=rf, param_grid=rf_grid, cv=5, scoring="neg_mean_squared_error"
)

# Execute the grid reserch
grid_search.fit(X, y)
```

```python
## Display the hyperparameters of the best model (code) ##
print("Best parameters:", grid_search.best_params_)
print("Best estimator:", grid_search.best_estimator_)
```

```python
## Print the score of the best model (code)##

## Here, best_score_ gives the Negative Mean Squared Error (NMSE),
## because the Scikit-Learn convention is that "bigger is better" for scoring functions.
## So, we need to take the negative value and square root it
## to get the Root Mean Squared Error (RMSE).
best_score = grid_search.best_score_
best_rmse = np.sqrt(-best_score)
print("Best RMSE:", best_rmse)
```
